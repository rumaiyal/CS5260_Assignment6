#Getting average of 3 runs , obtained for a few scenarios

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from os.path import exists
```

```python
os.path.expanduser('~')
```

```python
dirresultmain=os.path.join(os.path.expanduser('~'),'CS5260','Assignment6','Results')
fname_testacc='TestAccuracyVsStep'
fname_LR='LRVsStep'
fname_testloss='TestLossVsStep'
fname_trnloss='TrainLossVsStep'

fname_meantestacc='MeanTestAccuracyVsStep'
fname_meantestloss='MeanTestLossVsStep'
fname_meantrnloss='MeanTrainLossVsStep'
```

```python
##################
#Computing mean of 3 runs
###################
```

```python
#subdirname='SGD_Onecycle
#subdirname='SGD_FixedLR'
subdirname='LARS_WarmupLR'

resultpathfull_It1=os.path.join(dirresultmain,subdirname,'Iteration1')
resultpathfull_It2=os.path.join(dirresultmain,subdirname,'Iteration2')
resultpathfull_It3=os.path.join(dirresultmain,subdirname,'Iteration3')
resultpathfull_mean=os.path.join(dirresultmain,subdirname)

testacc_It1_fname_fpath=os.path.join(resultpathfull_It1,fname_testacc)
testacc_It2_fname_fpath=os.path.join(resultpathfull_It2,fname_testacc)
testacc_It3_fname_fpath=os.path.join(resultpathfull_It3,fname_testacc)
meantestacc_fname_fpath=os.path.join(resultpathfull_mean,fname_meantestacc)

testloss_It1_fname_fpath=os.path.join(resultpathfull_It1,fname_testloss)
testloss_It2_fname_fpath=os.path.join(resultpathfull_It2,fname_testloss)
testloss_It3_fname_fpath=os.path.join(resultpathfull_It3,fname_testloss)
meantestloss_fname_fpath=os.path.join(resultpathfull_mean,fname_meantestloss)

trnloss_It1_fname_fpath=os.path.join(resultpathfull_It1,fname_trnloss)
trnloss_It2_fname_fpath=os.path.join(resultpathfull_It2,fname_trnloss)
trnloss_It3_fname_fpath=os.path.join(resultpathfull_It3,fname_trnloss)
meantrnloss_fname_fpath=os.path.join(resultpathfull_mean,fname_meantrnloss)                          

dftestacc_It1=pd.read_csv(testacc_It1_fname_fpath)
dftestacc_It2=pd.read_csv(testacc_It2_fname_fpath)
dftestacc_It3=pd.read_csv(testacc_It3_fname_fpath)

dftestloss_It1=pd.read_csv(testloss_It1_fname_fpath)
dftestloss_It2=pd.read_csv(testloss_It2_fname_fpath)
dftestloss_It3=pd.read_csv(testloss_It3_fname_fpath)

dftrnloss_It1=pd.read_csv(trnloss_It1_fname_fpath)
dftrnloss_It2=pd.read_csv(trnloss_It2_fname_fpath)
dftrnloss_It3=pd.read_csv(trnloss_It3_fname_fpath)

dfmeantestacc=dftestacc_It1
dfmeantestloss=dftestloss_It1
dfmeantrnloss=dftrnloss_It1

dfmeantestacc.Value=(dftestacc_It1.Value+dftestacc_It2.Value+dftestacc_It3.Value)/3.0
dfmeantestloss.Value=(dftestloss_It1.Value+dftestloss_It2.Value+dftestloss_It3.Value)/3.0
dfmeantrnloss.Value=(dftrnloss_It1.Value+dftrnloss_It2.Value+dftrnloss_It3.Value)/3.0


plt.plot(dftestacc_It1.Step,dfmeantestacc.Value,color='black',label='Mean',linewidth=3)
plt.plot(dftestacc_It1.Step,dftestacc_It1.Value,color='red',label='It 1',linewidth=1)
plt.plot(dftestacc_It2.Step,dftestacc_It2.Value,color='blue',label='It 2',linewidth=1)
plt.plot(dftestacc_It3.Step,dftestacc_It3.Value,color='green',label='It 3',linewidth=1)
plt.xlabel('Step')
plt.ylabel('Mean Test Accuracy')
plt.grid()
plt.legend()
plt.show()

plt.plot(dftestacc_It1.Step,dfmeantestloss.Value,color='black',label='Mean',linewidth=3)
plt.plot(dftestacc_It1.Step,dftestloss_It1.Value,color='red',label='It 1',linewidth=1)
plt.plot(dftestacc_It2.Step,dftestloss_It2.Value,color='blue',label='It 2',linewidth=1)
plt.plot(dftestacc_It3.Step,dftestloss_It3.Value,color='green',label='It 3',linewidth=1)
plt.xlabel('Step')
plt.ylabel('Mean Test Loss')
plt.grid()
plt.legend()
plt.show()

plt.plot(dftrnloss_It1.Step,dfmeantrnloss.Value,color='black',label='Mean',linewidth=3)
plt.plot(dftrnloss_It1.Step,dftrnloss_It1.Value,color='red',label='It 1',linewidth=1)
plt.plot(dftrnloss_It2.Step,dftrnloss_It2.Value,color='blue',label='It 2',linewidth=1)
plt.plot(dftrnloss_It3.Step,dftrnloss_It3.Value,color='green',label='It 3',linewidth=1)
plt.xlabel('Step')
plt.ylabel('Mean Train Loss')
plt.grid()
plt.legend()
plt.show()



dfmeantestacc.to_csv(meantestacc_fname_fpath)
dfmeantestloss.to_csv(meantestloss_fname_fpath)
dfmeantrnloss.to_csv(meantrnloss_fname_fpath)
```
