```python
subdirname='SGD_FixedLR'

resultpathfull_mean_sgdfixedlr=os.path.join(dirresultmain,subdirname)
meantestacc_fname_fpath_sgdfixedlr=os.path.join(resultpathfull_mean_sgdfixedlr,fname_meantestacc)
meantestloss_fname_fpath_sgdfixedlr=os.path.join(resultpathfull_mean_sgdfixedlr,fname_meantestloss)
meantrnloss_fname_fpath_sgdfixedlr=os.path.join(resultpathfull_mean_sgdfixedlr,fname_meantrnloss)                          

dfmeantestacc_sgdfixedlr=pd.read_csv(meantestacc_fname_fpath_sgdfixedlr)
dfmeantestloss_sgdfixedlr=pd.read_csv(meantestloss_fname_fpath_sgdfixedlr)
dfmeantrnloss_sgdfixedlr=pd.read_csv(meantrnloss_fname_fpath_sgdfixedlr)


subdirname='SGD_MultiStepLR'

resultpathfull_sgdmultisteplr=os.path.join(dirresultmain,subdirname)
testacc_fname_fpath_sgdmultisteplr=os.path.join(resultpathfull_sgdmultisteplr,fname_testacc)
testloss_fname_fpath_sgdmultisteplr=os.path.join(resultpathfull_sgdmultisteplr,fname_testloss)
trnloss_fname_fpath_sgdmultisteplr=os.path.join(resultpathfull_sgdmultisteplr,fname_trnloss)                          

dftestacc_sgdmultisteplr=pd.read_csv(testacc_fname_fpath_sgdmultisteplr)
dftestloss_sgdmultisteplr=pd.read_csv(testloss_fname_fpath_sgdmultisteplr)
dftrnloss_sgdmultisteplr=pd.read_csv(trnloss_fname_fpath_sgdmultisteplr)


subdirname='SGD_Onecycle'

resultpathfull_mean_sgdonecycle=os.path.join(dirresultmain,subdirname)
meantestacc_fname_fpath_sgdonecycle=os.path.join(resultpathfull_mean_sgdonecycle,fname_meantestacc)
meantestloss_fname_fpath_sgdonecycle=os.path.join(resultpathfull_mean_sgdonecycle,fname_meantestloss)
meantrnloss_fname_fpath_sgdonecycle=os.path.join(resultpathfull_mean_sgdonecycle,fname_meantrnloss)                          

dfmeantestacc_sgdonecycle=pd.read_csv(meantestacc_fname_fpath_sgdonecycle)
dfmeantestloss_sgdonecycle=pd.read_csv(meantestloss_fname_fpath_sgdonecycle)
dfmeantrnloss_sgdonecycle=pd.read_csv(meantrnloss_fname_fpath_sgdonecycle)


plt.plot(dfmeantestacc_sgdfixedlr.Step,dfmeantestacc_sgdfixedlr.Value,color='red',label='Mean SGD With Optimally Fixed LR=0.13')
plt.plot(dftestacc_sgdmultisteplr.Step,dftestacc_sgdmultisteplr.Value,color='blue',label='SGD With Step LR, Start LR=0.13, epoch step size=10, decay=0.1',linestyle='dashed')
plt.plot(dfmeantestacc_sgdonecycle.Step,dfmeantestacc_sgdonecycle.Value,color='green',label='Mean SGD With One Cycle LR, Max LR=0.32, Init Div Factor=8')
plt.xlabel('Step')
plt.ylabel('Mean\Sample Test Accuracy')
plt.grid()
plt.legend()
plt.show()

#plt.plot(dfmeantestloss_sgdfixedlr.Step,dfmeantestloss_sgdfixedlr.Value,color='red',label='Mean SGD With Optimally Fixed LR=0.13')
#plt.plot(dftestloss_sgdmultisteplr.Step,dftestloss_sgdmultisteplr.Value,color='blue',label='SGD With Step LR, Start LR=0.13, epoch step size=10, decay=0.1',linestyle='dashed')
#plt.plot(dfmeantestloss_sgdonecycle.Step,dfmeantestloss_sgdonecycle.Value,color='green',label='Mean SGD With One Cycle LR, Max LR=0.32, Init Div Factor=8')
#plt.xlabel('Step')
#plt.ylabel('Mean\Sample Test Loss')
#plt.grid()
#plt.legend()
#plt.show()

plt.plot(dfmeantrnloss_sgdfixedlr.Step,dfmeantrnloss_sgdfixedlr.Value,color='red',label='Mean SGD With Optimally Fixed LR=0.13')
plt.plot(dftrnloss_sgdmultisteplr.Step,dftrnloss_sgdmultisteplr.Value,color='blue',label='SGD With Step LR, Start LR=0.13, epoch step size=10, decay=0.1',linestyle='dashed')
plt.plot(dfmeantrnloss_sgdonecycle.Step,dfmeantrnloss_sgdonecycle.Value,color='green',label='Mean SGD With One Cycle LR, Max LR=0.32, Init Div Factor=8')
plt.xlabel('Step')
plt.ylabel('Mean\Sample Train Loss')
plt.grid()
plt.legend()
plt.show()
```
