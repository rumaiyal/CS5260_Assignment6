```python
dirresultmain=os.path.join(os.path.expanduser('~'),'CS5260','Assignment6','Results')
fname_testacc='TestAccuracyVsStep'
fname_LR='LRVsStep'
fname_testloss='TestLossVsStep'
fname_trnloss='TrainLossVsStep'

fname_meantestacc='MeanTestAccuracyVsStep'
fname_meantestloss='MeanTestLossVsStep'
fname_meantrnloss='MeanTrainLossVsStep'
```

```python
subdirname='SGD_FixedLR'

resultpathfull_mean_sgdfixedlr=os.path.join(dirresultmain,subdirname)
meantestacc_fname_fpath_sgdfixedlr=os.path.join(resultpathfull_mean_sgdfixedlr,fname_meantestacc)
meantestloss_fname_fpath_sgdfixedlr=os.path.join(resultpathfull_mean_sgdfixedlr,fname_meantestloss)
meantrnloss_fname_fpath_sgdfixedlr=os.path.join(resultpathfull_mean_sgdfixedlr,fname_meantrnloss)                          

dfmeantestacc_sgdfixedlr=pd.read_csv(meantestacc_fname_fpath_sgdfixedlr)
dfmeantestloss_sgdfixedlr=pd.read_csv(meantestloss_fname_fpath_sgdfixedlr)
dfmeantrnloss_sgdfixedlr=pd.read_csv(meantrnloss_fname_fpath_sgdfixedlr)


subdirname='Adam_FixedLR'

resultpathfull_adamfixedlr=os.path.join(dirresultmain,subdirname)
testacc_fname_fpath_adamfixedlr=os.path.join(resultpathfull_adamfixedlr,fname_testacc)
testloss_fname_fpath_adamfixedlr=os.path.join(resultpathfull_adamfixedlr,fname_testloss)
trnloss_fname_fpath_adamfixedlr=os.path.join(resultpathfull_adamfixedlr,fname_trnloss)                          

dftestacc_adamfixedlr=pd.read_csv(testacc_fname_fpath_adamfixedlr)
dftestloss_adamfixedlr=pd.read_csv(testloss_fname_fpath_adamfixedlr)
dftrnloss_adamfixedlr=pd.read_csv(trnloss_fname_fpath_adamfixedlr)


subdirname='LARS_WarmupLR'

resultpathfull_mean_larswarmuplr=os.path.join(dirresultmain,subdirname)
meantestacc_fname_fpath_larswarmuplr=os.path.join(resultpathfull_mean_larswarmuplr,fname_meantestacc)
meantestloss_fname_fpath_larswarmuplr=os.path.join(resultpathfull_mean_larswarmuplr,fname_meantestloss)
meantrnloss_fname_fpath_larswarmuplr=os.path.join(resultpathfull_mean_larswarmuplr,fname_meantrnloss)                          

dfmeantestacc_larswarmuplr=pd.read_csv(meantestacc_fname_fpath_larswarmuplr)
dfmeantestloss_larswarmuplr=pd.read_csv(meantestloss_fname_fpath_larswarmuplr)
dfmeantrnloss_larswarmuplr=pd.read_csv(meantrnloss_fname_fpath_larswarmuplr)


plt.plot(dfmeantestacc_sgdfixedlr.Step,dfmeantestacc_sgdfixedlr.Value,color='red',label='Mean SGD With Optimally Fixed LR=0.13')
plt.plot(dftestacc_adamfixedlr.Step,dftestacc_adamfixedlr.Value,color='blue',label='Adam With Fixed LR, Start LR=0.13, epoch step size=10, decay=0.1',linestyle='dashed')
plt.plot(dfmeantestacc_larswarmuplr.Step,dfmeantestacc_larswarmuplr.Value,color='green',label='LARS With Warmup, Polynomial Decay, Same params as Imagenet')
plt.xlabel('Step')
plt.ylabel('Mean\Sample Test Accuracy')
plt.grid()
plt.legend()
plt.show()

#plt.plot(dfmeantestloss_sgdfixedlr.Step,dfmeantestloss_sgdfixedlr.Value,color='red',label='Mean SGD With Optimally Fixed LR=0.13')
#plt.plot(dftestloss_sgdmultisteplr.Step,dftestloss_sgdmultisteplr.Value,color='blue',label='SGD With Step LR, Start LR=0.13, epoch step size=10, decay=0.1',linestyle='dashed')
#plt.plot(dfmeantestloss_sgdonecycle.Step,dfmeantestloss_sgdonecycle.Value,color='green',label='Mean SGD With One Cycle LR, Max LR=0.32, Init Div Factor=8')
#plt.xlabel('Step')
#plt.ylabel('Mean\Sample Test Loss')
#plt.grid()
#plt.legend()
#plt.show()

plt.plot(dfmeantrnloss_sgdfixedlr.Step,dfmeantrnloss_sgdfixedlr.Value,color='red',label='Mean SGD With Optimally Fixed LR=0.13')
plt.plot(dftrnloss_adamfixedlr.Step,dftrnloss_adamfixedlr.Value,color='blue',label='Adam With Fixed LR, Start LR=0.13, epoch step size=10, decay=0.1',linestyle='dashed')
plt.plot(dfmeantrnloss_larswarmuplr.Step,dfmeantrnloss_larswarmuplr.Value,color='green',label='LARS With Warmup, Polynomial Decay, Same params as Imagenet')
plt.xlabel('Step')
plt.ylabel('Mean\Sample Train Loss')
plt.grid()
plt.legend()
plt.show()
```
