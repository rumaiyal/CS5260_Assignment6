#Plotting First Pass and Second Fine tuning pass to get Optimal LR from Super Convergence

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from os.path import exists
```

```python
os.path.expanduser('~')
```

```python
dirresultmain=os.path.join(os.path.expanduser('~'),'CS5260','Assignment6','Results')
fname_testacc='TestAccuracyVsStep'
fname_LR='LRVsStep'
fname_testloss='TestLossVsStep'
fname_trnloss='TrainLossVsStep'
```

```python
subdirname='SGD_superconverge'

resultpathfull_sgdsuperconverge=os.path.join(dirresultmain,subdirname)
trnloss_fname_fpath_sgdsuperconverge=os.path.join(resultpathfull_sgdsuperconverge,fname_trnloss)  
lr_fname_fpath_sgdsuperconverge=os.path.join(resultpathfull_sgdsuperconverge,fname_LR)

dftrnloss_sgdsuperconverge=pd.read_csv(trnloss_fname_fpath_sgdsuperconverge)
dflr_sgdsuperconverge=pd.read_csv(lr_fname_fpath_sgdsuperconverge)


subdirname='SGD_superconverge_finetune'

resultpathfull_sgdsuperconvergefinetune=os.path.join(dirresultmain,subdirname)
trnloss_fname_fpath_sgdsuperconvergefinetune=os.path.join(resultpathfull_sgdsuperconvergefinetune,fname_trnloss)  
lr_fname_fpath_sgdsuperconvergefinetune=os.path.join(resultpathfull_sgdsuperconvergefinetune,fname_LR)

dftrnloss_sgdsuperconvergefinetune=pd.read_csv(trnloss_fname_fpath_sgdsuperconvergefinetune)
dflr_sgdsuperconvergefinetune=pd.read_csv(lr_fname_fpath_sgdsuperconvergefinetune)


plt.scatter(dflr_sgdsuperconverge.Value,dftrnloss_sgdsuperconverge.Value,color='red',marker='.',s=10,label='Super Convergence with SGD, after first pass')
#plt.scatter(dflr_sgdsuperconvergefinetune.Value,dftrnloss_sgdsuperconvergefinetune.Value,color='blue',marker='.',s=10,label='Super Convergence with SGD, after second fine-tune pass')
plt.xlabel('Learning Rate')
plt.ylabel('Train Loss')
plt.ylim([0.0,3.0])
plt.grid()
plt.legend()
plt.show()

```
