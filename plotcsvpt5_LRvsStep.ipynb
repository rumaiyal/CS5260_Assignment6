#Plotting First Pass and Second Fine tuning pass to get Optimal LR from Super Convergence

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from os.path import exists
```

```python
os.path.expanduser('~')
```

```python
dirresultmain=os.path.join(os.path.expanduser('~'),'CS5260','Assignment6','Results')
fname_testacc='TestAccuracyVsStep'
fname_LR='LRVsStep'
fname_testloss='TestLossVsStep'
fname_trnloss='TrainLossVsStep'
```

```python
subdirname='SGD_FixedLR'

resultpathfull_sgdfixedlr=os.path.join(dirresultmain,subdirname,'Iteration1')
lr_fname_fpath_sgdfixedlr=os.path.join(resultpathfull_sgdfixedlr,fname_LR)
dflr_sgdfixedlr=pd.read_csv(lr_fname_fpath_sgdfixedlr)


subdirname='SGD_MultiStepLR'

resultpathfull_sgdsteplr=os.path.join(dirresultmain,subdirname)
lr_fname_fpath_sgdsteplr=os.path.join(resultpathfull_sgdsteplr,fname_LR)
dflr_sgdsteplr=pd.read_csv(lr_fname_fpath_sgdsteplr)

subdirname='SGD_Onecycle'

resultpathfull_sgdonecycle=os.path.join(dirresultmain,subdirname,'Iteration1')  
lr_fname_fpath_sgdonecycle=os.path.join(resultpathfull_sgdonecycle,fname_LR)
dflr_sgdonecycle=pd.read_csv(lr_fname_fpath_sgdonecycle)

plt.plot(dflr_sgdfixedlr.Step,dflr_sgdfixedlr.Value,color='red',label='Fixed Learning rate')
plt.plot(dflr_sgdsteplr.Step,dflr_sgdsteplr.Value,color='blue',label='Stepped Learning rate')
plt.plot(dflr_sgdonecycle.Step,dflr_sgdonecycle.Value,color='green',label='One Cycle Learning rate')

plt.xlabel('Step')
plt.ylabel('Learning Rate')
plt.grid()
plt.legend()
plt.show()

```

```python
subdirname='LARS_WarmupLR'

resultpathfull_larswarmuplr=os.path.join(dirresultmain,subdirname,'Iteration1')
lr_fname_fpath_larswarmuplr=os.path.join(resultpathfull_larswarmuplr,fname_LR)
dflr_larswarmuplr=pd.read_csv(lr_fname_fpath_larswarmuplr)
```

```python
plt.plot(dflr_larswarmuplr.Step,dflr_larswarmuplr.Value,color='red',label='LARS Learning rate, with warm up')

plt.xlabel('Step')
plt.ylabel('Learning Rate')
plt.grid()
plt.legend()
plt.show()
```
